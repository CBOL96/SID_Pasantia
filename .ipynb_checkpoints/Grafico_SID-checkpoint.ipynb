```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path='C:/Users/Carlos/Documents/Pasantia/SID_Pasantia/Datos/'
mediciones=pd.read_csv(path+"CITEC_VTX3_2021-02-26.csv",names=['Tiempo UTC','Amplitud'],dtype={'Tiempo UTC':'str','Amplitud':'float64'})
datos=pd.DataFrame(mediciones)
```

```python
datos['Tiempo UTC']=pd.to_datetime(datos['Tiempo UTC'])
datos.set_index('Tiempo UTC',drop=True, inplace=True)
datos.head()
flag=datos.index.values
```

```python
fig,axes = plt.subplots(1,1, figsize=(20,6),sharex=True)
axes=datos['Amplitud'][flag].plot(ylim=(0,120),legend=False)
ax=plt.gca()
ax.set_ylabel('Intensidad de la Señal')
plt.show()
```
